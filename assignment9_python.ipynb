## Problem 1(a).  College data.

The data set `usnews.csv` in the folder with this notebook contains data from the 1995 US News and World Report on colleges and universities in the US.
* (Dataset:  ‚ÄúUS News College data,‚Äù submitted by Robin Lock, St. Lawrence University.   Dataset obtained from the Journal of Statistics Education (http://www.amstat.org/publications/jse). Accessed 23 June 2015. Used by permission of author.)
* Metadata:  http://www.amstat.org/publications/jse/datasets/usnews.txt


The data set contains asterisks which indicate missing data.  

üéØ Use pandas to create a data frame which replaces the asterisks by ```numpy.nan```. Also look out for extra spaces.  

````python
#Import libraries
import numpy as np
import pandas as pd

#Open File with Panda File
#Take the spaces off the leading and trailing header columns
def read_file(in_name):
    out_data = pd.read_csv(in_name)
#    out_data = [c.strip() for c in out_data.columns]  
    
    return out_data

#Replace Asterisks in the data with numpy.nan
def replace_asterisks(in_data):
    if in_data == "*":
        return np.nan
    else:
        return in_data

data_set = read_file('usnews.csv')
data_set = data_set.applymap(replace_asterisks)
data_set.head(n=5)
#data_set.describe()
````

## Problem 1(b).  Categorizing universities.

üéØ Create a new column in your data frame which describes whether a university is public or private, using the words "Public" and "Private" instead of numbers.  (It may be helpful to consult the metadata file.)


````python
def add_column(in_data):    
    in_data['Public_Private_Descr'] = np.where(in_data[['Public/private']]==1,'Public','Private')
    return in_data

data_set_2 = add_column(data_set)
data_set_2.head(5)
#print(data_set_2[['Public/private','Public_Private_Descr']])
````

## Problem 1(c).  

The interquartile range is the 3rd quartile minus the 1st quartile. 

üéØ Create columns recording the interquartile range for math and verbal SAT scores.

````python
def interquartile(in_data, field_name_q1, field_name_q3, new_name):    
    Third_q = in_data[field_name_q3].factorize()[0]
    First_q = in_data[field_name_q1].factorize()[0]
    new_data = Third_q - First_q
    in_data[new_name] = new_data 
    return in_data

data_set_3 = interquartile(data_set_2,'First quartile - Math SAT','Third quartile - Math SAT','Interquartile - Math SAT')

data_set_4 = interquartile(data_set_3,'First quartile - Verbal SAT','Third quartile - Verbal SAT','Interquartile - Verbal SAT')

data_set_4.head()
````

## Problem 1(d).  Saving for R.

üéØ Save your data frame as a `.csv` file for further analysis in R.  

* You must replace the ```nan```s with an empty string before saving.  
* Your output should not include a column of row numbers.  Omit the indices.

````python
#Write File 
def write_file(in_name,in_data):
    in_data.fillna("", inplace=True)
    in_data.to_csv(in_name, sep=',',index=False)
    return

write_file('assignment9_out.csv',data_set_4)
````
